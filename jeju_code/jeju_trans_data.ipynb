```python
import os
from tqdm.notebook import tqdm

df = os.listdir('/content/valid_train_text_file')
```

```python
a=[]
for i in tqdm(df):
  with open(f"/content/valid_train_text_file/{i}", "r") as f:
      for line in f:
        if('(' in line):
          a.append(line.strip())
```

```python
dic = {
    'irregular' : [],
    'standard' : []

}
for k in tqdm(range(len(a))):
  x_sent = a[k].split(' ')
  y_sent = a[k].split(' ')
  for j,i in enumerate(a[k].split(' ')):
    if('(' in i):
      if( len(i.split('/')) == 2):
        x,y = i.split('/')
      
      x_sent[j] = x[1:-1]
      y_sent[j] = y[1:-1]
  dic['irregular'].append(' '.join(x_sent[1:]))
  dic['standard'].append(' '.join(y_sent[1:]))


    
```

```python
import pandas as pd

df = pd.DataFrame(dic)
df
```

```python
df = df.dropna(axis = 0)
```
